```python
dataset = []
for i in range(1000):
    a = list(" ".join(line[i].split()))
    memory = ""
    main = []
    for i in a:
        if i != " ":
            memory+=i
        else:
            main.append(int(memory))
            memory = " "
    if memory != " ":
        main.append(int(memory))
    dataset.append(main)
```

```python
## logistic regression

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from subprocess import check_output
```

```python
a = 'a'
name=[]
for i in range(25):
    name.append(a+str(i+1))
df.columns = name
```

```python
#Test set train set 
test_set = df.sample(n=100)
print(test_set.head())
len(test_set)


train_set = pd.concat([df, test_set, test_set]).drop_duplicates(keep=False)
```

```python
X = train_set.loc[:,['a1', 'a2', 'a3']]
Y = train_set.loc[:,'a25']
Xtest = test_set.loc[:,['a1', 'a2', 'a3']]
```

```python
print(len(X))
print(len(Y))
print(len(Xtest))
```

```python
Xtrain,Xvalid,Ytrain,Yvalid = train_test_split(X,Y
                                               , test_size=0.33, random_state = 42)
```

```python
LR = LogisticRegression(C=10)
LR.fit(X=Xtrain, y=Ytrain)
```

```python
yPredTrain=LR.predict(Xtrain)
yPredValid = LR.predict(Xvalid)
```

```python
def acc(Y,yPred):
    return np.sum(yPred==Y)/len(Y)
```

```python
print('Train set accuracy', acc(Ytrain, yPredTrain), '%')
print('Validation set accuracy', acc(Yvalid, yPredValid), '%')
```
